```python
#Across locations, how many apples were purchased in cash?
apples_cash = df[(df["Product Name"].str.lower() == "apple") & 
                 (df["Payment Method"].str.lower() == "cash")]

total_apples_cash = apples_cash["Quantity"].sum()
print("Total apples purchased in cash:", total_apples_cash)

#How much total cash was spent on these apples?
total_cash_spent_on_apples = apples_cash["Total Amount"].sum()
print("Total cash spent on apples:", total_cash_spent_on_apples)

#Across all payment methods, how much money was spent at the Bakershire store by non-member customers?
bakershire_non_members = df[(df["Store"].str.lower() == "bakershire") & 
                            (df["Customer Type"].str.lower() == "non-member")]

total_spent_bakershire_non_members = bakershire_non_members["Total Amount"].sum()
print("Total spent at Bakershire by non-members:", total_spent_bakershire_non_members)

```

```python
#Group by Product + Payment Method
grouped = df.groupby(["Product Name", "Payment Method"]).agg(
    Total_Quantity=("Quantity", "sum"),
    Total_Spent=("Total Amount", "sum")
).reset_index()

print(grouped)

#Group by Store + Customer Type

store_summary = df.groupby(["Store", "Customer Type"]).agg(
    Total_Spent=("Total Amount", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

print(store_summary)

```

```python
#Combine Filters + Groupby (flexible analysis)

apples_cash = df[(df["Product Name"].str.lower() == "apple") & 
                 (df["Payment Method"].str.lower() == "cash")]

summary_apples_cash = apples_cash.groupby("Store").agg(
    Total_Quantity=("Quantity", "sum"),
    Total_Spent=("Total Amount", "sum")
).reset_index()

print(summary_apples_cash)

#Pivot Table Style (Excel-like view)

pivot = pd.pivot_table(
    df,
    values="Total Amount",
    index=["Store"],
    columns=["Customer Type", "Payment Method"],
    aggfunc="sum",
    fill_value=0
)

print(pivot)
```
